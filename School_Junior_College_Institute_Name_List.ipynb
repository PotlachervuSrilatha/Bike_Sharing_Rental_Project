```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Replace this URL with the actual website you want to scrape
url = 'https://www.enrolmentdesk.com/junior-colleges'

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the relevant elements on the page and extract data
    # Replace these lines with the actual HTML elements of the page you want to scrape
    names = [element.text for element in soup.find_all('h3', class_='school-name')]
    locations = [element.text for element in soup.find_all('p', class_='school-location')]
    contact_numbers = [element.text for element in soup.find_all('span', class_='contact-number')]
    website_urls = [element['href'] for element in soup.find_all('a', class_='school-link')]

    # Create a DataFrame using pandas
    data = {
        'School/Junior College/Institute': names,
        'Location': locations,
        'Contact Number': contact_numbers,
        'Institute Website URL': website_urls
    }
    df = pd.DataFrame(data)

    # Display the DataFrame or save it to an Excel file
    print(df)

    # Save to Excel file
    df.to_excel('junior_colleges_list.xlsx', index=False)

```
